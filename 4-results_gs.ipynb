```python
import numpy as np
import pandas as pd
```

```python
res_cefta = pd.read_csv("results/grid_search/cv_results__Ceftazidim.csv")
res_cipro = pd.read_csv("results/grid_search/cv_results__Ciprofloxacin.csv")
res_colis = pd.read_csv("results/grid_search/cv_results__Colistin.csv")
res_merop = pd.read_csv("results/grid_search/cv_results__Meropenem.csv")
res_tobra = pd.read_csv("results/grid_search/cv_results__Tobramycin.csv")
```

```python
cols = ["mean_test_score", "std_test_score", "param_sel_ind__gpa", "param_sel_ind__snps", "param_sel_ind__genexp",
        "param_dim_red", "param_clf"]
best_cefta = res_cefta.sort_values(["mean_test_score", "std_test_score", "rank_test_score"],
                                   axis=0, ascending=[False, True, True]).loc[:, cols].iloc[0:1,]
best_cefta.index = ["Ceftazidim"]
best_cipro = res_cipro.sort_values(["mean_test_score", "std_test_score", "rank_test_score"],
                                   axis=0, ascending=[False, True, True]).loc[:, cols].iloc[0:1,]
best_cipro.index = ["Ciprofloxacin"]
best_colis = res_colis.sort_values(["mean_test_score", "std_test_score", "rank_test_score"],
                                   axis=0, ascending=[False, True, True]).loc[:, cols].iloc[0:1,]
best_colis.index = ["Colistin"]
best_merop = res_merop.sort_values(["mean_test_score", "std_test_score", "rank_test_score"],
                                   axis=0, ascending=[False, True, True]).loc[:, cols].iloc[0:1,]
best_merop.index = ["Meropenem"]
best_tobra = res_tobra.sort_values(["mean_test_score", "std_test_score", "rank_test_score"],
                                   axis=0, ascending=[False, True, True]).loc[:, cols].iloc[0:1,]
best_tobra.index = ["Tobramycin"]

res = pd.concat([best_cefta, best_cipro, best_colis, best_merop, best_tobra], axis=0)

def replace_sel_ind(x):
    return x == "passthrough"
res["param_sel_ind__gpa"] = res["param_sel_ind__gpa"].apply(replace_sel_ind)
res["param_sel_ind__snps"] = res["param_sel_ind__snps"].apply(replace_sel_ind)
res["param_sel_ind__genexp"] = res["param_sel_ind__genexp"].apply(replace_sel_ind)

def crop_est(x):
    return x.split("(")[0]
res["param_dim_red"] = res["param_dim_red"].apply(crop_est)
res["param_clf"] = res["param_clf"].apply(crop_est)

columns = pd.MultiIndex.from_tuples(list(zip(["Balanced accuracy"]*2, ["moyenne", "déviation"])) +
                                    list(zip(["Présence"]*3, ["gpa", "snps", "genexp"])) +
                                    list(zip(["Estimateur"]*2, ["réduction", "classifieur"])))
res = res.set_axis(columns, axis=1)

res.to_latex(buf="rapport/tables/res_cv.tex",
            float_format="%.2f",
            bold_rows=True,
            escape=True,
            caption="Meilleurs scores obtenus par validation croisée et architectures associées.",
            label="tab:res_cv",
            position="H")
with open("rapport/tables/res_cv.tex", "r+") as f:
    text = f.read().replace("\\begin{tabular}", "\\resizebox{\\textwidth}{!}{\\begin{tabular}")
    text = text.replace("\\end{tabular}", "\\end{tabular}}")
    text = text.replace("Balanced accuracy", "\\textit{Balanced accuracy}")
    f.seek(0)
    f.write(text)
    f.close()
```

```python
cols = ["param_sel_ind__gpa", "param_sel_ind__snps", "param_sel_ind__genexp"]

best_cefta = res_cefta.sort_values(["mean_test_score", "std_test_score", "rank_test_score"],
                                   axis=0, ascending=[False, True, True]).loc[:, cols].iloc[0:10,]
best_cefta["param_sel_ind__gpa"] = best_cefta["param_sel_ind__gpa"].apply(replace_sel_ind)
best_cefta["param_sel_ind__snps"] = best_cefta["param_sel_ind__snps"].apply(replace_sel_ind)
best_cefta["param_sel_ind__genexp"] = best_cefta["param_sel_ind__genexp"].apply(replace_sel_ind)
best_cefta = pd.DataFrame(best_cefta.aggregate("sum", axis=0).to_numpy(),
                          index=["gpa", "snps", "genexp"], columns=["Ceftazidim"]).T
best_cipro = res_cipro.sort_values(["mean_test_score", "std_test_score", "rank_test_score"],
                                   axis=0, ascending=[False, True, True]).loc[:, cols].iloc[0:10,]
best_cipro["param_sel_ind__gpa"] = best_cipro["param_sel_ind__gpa"].apply(replace_sel_ind)
best_cipro["param_sel_ind__snps"] = best_cipro["param_sel_ind__snps"].apply(replace_sel_ind)
best_cipro["param_sel_ind__genexp"] = best_cipro["param_sel_ind__genexp"].apply(replace_sel_ind)
best_cipro = pd.DataFrame(best_cipro.aggregate("sum", axis=0).to_numpy(),
                          index=["gpa", "snps", "genexp"], columns=["Ciprofloxacin"]).T
best_colis = res_colis.sort_values(["mean_test_score", "std_test_score", "rank_test_score"],
                                   axis=0, ascending=[False, True, True]).loc[:, cols].iloc[0:10,]
best_colis["param_sel_ind__gpa"] = best_colis["param_sel_ind__gpa"].apply(replace_sel_ind)
best_colis["param_sel_ind__snps"] = best_colis["param_sel_ind__snps"].apply(replace_sel_ind)
best_colis["param_sel_ind__genexp"] = best_colis["param_sel_ind__genexp"].apply(replace_sel_ind)
best_colis = pd.DataFrame(best_colis.aggregate("sum", axis=0).to_numpy(),
                          index=["gpa", "snps", "genexp"], columns=["Colistin"]).T
best_merop = res_merop.sort_values(["mean_test_score", "std_test_score", "rank_test_score"],
                                   axis=0, ascending=[False, True, True]).loc[:, cols].iloc[0:10,]
best_merop["param_sel_ind__gpa"] = best_merop["param_sel_ind__gpa"].apply(replace_sel_ind)
best_merop["param_sel_ind__snps"] = best_merop["param_sel_ind__snps"].apply(replace_sel_ind)
best_merop["param_sel_ind__genexp"] = best_merop["param_sel_ind__genexp"].apply(replace_sel_ind)
best_merop = pd.DataFrame(best_merop.aggregate("sum", axis=0).to_numpy(),
                          index=["gpa", "snps", "genexp"], columns=["Meropenem"]).T
best_tobra = res_tobra.sort_values(["mean_test_score", "std_test_score", "rank_test_score"],
                                   axis=0, ascending=[False, True, True]).loc[:, cols].iloc[0:10,]
best_tobra["param_sel_ind__gpa"] = best_tobra["param_sel_ind__gpa"].apply(replace_sel_ind)
best_tobra["param_sel_ind__snps"] = best_tobra["param_sel_ind__snps"].apply(replace_sel_ind)
best_tobra["param_sel_ind__genexp"] = best_tobra["param_sel_ind__genexp"].apply(replace_sel_ind)
best_tobra = pd.DataFrame(best_tobra.aggregate("sum", axis=0).to_numpy(),
                          index=["gpa", "snps", "genexp"], columns=["Tobramycin"]).T

res = pd.concat([best_cefta, best_cipro, best_colis, best_merop, best_tobra], axis=0)
res = res.apply(lambda x: (x*10).astype(str) + "%", axis=1)

res.to_latex(buf="rapport/tables/res_x.tex",
            float_format="%.2f",
            bold_rows=True,
            escape=True,
            caption="Pourcentage de sélection des matrices de régresseurs parmi les 10 meilleures architectures.",
            label="tab:res_x",
            position="H")
```

```python
def merge_df(df1, df2):
    df = pd.concat([df1, df2], axis=0, ignore_index=True)
    df.iloc[np.argsort(-df["mean_test_score"].to_numpy()), -1] = np.arange(1, df.shape[0] + 1)
    return df
```
